```python
def remove_empty_lines(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        lines = f.readlines()
        for line in lines:
            if not line.strip():
                lines.remove(line)

    with open(file_path, "w", encoding="utf-8") as f:
        for line in lines:
            f.write(line)

remove_empty_lines("article.txt")
```

```python
def get_title(file_path):
    with open(file_path, "r") as file:
        title = file.readline()
    return title.strip()

get_title("article.txt")
```

```python
def get_text(file_path):
    with open(file_path, "r") as file:
        text = []
        for _ in range(6): 
            line = file.readline()
            if not line:  # Check if end of file is reached
                break
            text.append(line.strip())
        text = text[1:6]
        text = " ".join(text)
    return text
get_text("article.txt")
```

```python
import re

def get_code_snippets(content):
    
    codes = []
    pattern = re.compile(r'\d+\..*?(?=\n\d+\.|\bWorking\b)', re.DOTALL)
    matches = pattern.findall(content)
    for match in matches:
        # Remove section numbering and title
        code_snippet = re.sub(r'^\d+\..*?\n|([A-Z].*?:.*?\n)|(\n)+$', '', match)
        codes.append(code_snippet.strip())
    return codes
```

```python
def format_text(file_path):

    subtitle_pattern = r'^Working.*(?<!:)$'  # Pattern to match lines starting with 'Working'
    subtitle_item_pattern = r'^\d+.'


    with open(file_path, "r", encoding='utf-8') as file:
        content = file.read()
        codes = get_code_snippets(content)

        lines = content.split("\n")
        write_title(lines[0])
        write_text(" ".join(lines[1:6]))
        for line in lines[6:]:
            if re.match(subtitle_pattern, line):
                write_subtitle(line)
            elif re.match(subtitle_item_pattern, line):
                write_subtitle_item(line)
            elif line.strip() in codes:  # Check if the line is a code snippet
                write_code_snippet(line)
            else:
                write_comment(line)

format_text("article.txt")
```
